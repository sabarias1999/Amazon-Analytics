```python
import pandas as pd
```

```python
retail=pd.read_csv("E:/Amazon Analytics/Data/online_retail.csv")
```

```python
reviews=pd.read_csv("E:/Amazon Analytics/Data/amazon_reviews.csv")
```

```python
print(retail.head())
```

```python
print("RETAIL DATA:")
print(retail.head())
print(retail.info())

print("\nREVIEWS DATA:")
print(reviews.head())
print(reviews.info())
```

```python
products = retail.rename(columns={
    "asins": "product_id",
    "reviews.rating": "rating",
    "reviews.text": "review_text",
    "reviews.title": "review_title",
    "reviews.username": "customer_name"
})
```

```python
products = products[[
    "product_id",
    "brand",
    "categories",
    "rating",
    "review_text"
]]
```

```python
products = products.dropna(subset=["rating", "review_text"])
products.drop_duplicates(inplace=True)
print(products.head())
```

```python
reviews = reviews.rename(columns={
    "ProductId": "product_id",
    "Score": "rating",
    "Text": "review_text",
    "UserId": "customer_id",
    "Time": "review_time"
})
```

```python
reviews["review_date"] = pd.to_datetime(reviews["review_time"], unit='s')
```

```python
reviews = reviews[[
    "product_id",
    "customer_id",
    "rating",
    "review_text",
    "review_date"
]]
```

```python
reviews.drop_duplicates(inplace=True)
reviews = reviews.dropna(subset=["rating"])

print(reviews.head())
```

```python
products["customer_id"] = None
products["review_date"] = None


reviews["brand"] = None
reviews["categories"] = None


products = products[[
    "product_id", "customer_id", "rating", "review_text", "review_date", "brand", "categories"
]]

reviews = reviews[[
    "product_id", "customer_id", "rating", "review_text", "review_date", "brand", "categories"
]]
```

Merging

```python
df = pd.concat([products, reviews], ignore_index=True)

print(df.shape)
```

```python
df["review_length"] = df["review_text"].astype(str).apply(len)


def rating_label(x):
    if x >= 4:
        return "positive"
    elif x == 3:
        return "neutral"
    else:
        return "negative"

df["rating_category"] = df["rating"].apply(rating_label)
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

df["sentiment_score"] = df["review_text"].astype(str).apply(
    lambda x: sia.polarity_scores(x)["compound"]
)
```

```python
# Sentiment calculation
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

df["sentiment_score"] = df["review_text"].astype(str).apply(
    lambda x: sia.polarity_scores(x)["compound"]
)

def sentiment_label(x):
    if x > 0.05:
        return "positive"
    elif x < -0.05:
        return "negative"
    else:
        return "neutral"

df["sentiment"] = df["sentiment_score"].apply(sentiment_label)
```
